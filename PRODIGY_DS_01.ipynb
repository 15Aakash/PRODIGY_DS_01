### Data Visualization and Interpretation

```python
plt.figure(figsize=(8, 6))
sns.histplot(data["Age"], bins = 17, color = 'skyblue')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The histogram visually represents the distribution of customer ages, showing which age groups are more prevalent among the customers and providing insights into the age demographics of the customer base.

```python
plt.figure(figsize=(8, 6))
sns.histplot(data["Purchase Amount (USD)"], bins = 10, color = 'red' , edgecolor='black')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Amount (USD)')
plt.show()
```

The histogram provides insights into the distribution of purchase amounts among customers, showing which purchase amount ranges are more common and how frequently purchases occur within each range.

```python
plt.figure(figsize=(10, 6))
df['Category'].value_counts().plot(kind='bar', color='black')
plt.xlabel('Category')
plt.ylabel('Age')
plt.title('Age Distribution of Customers')
plt.xticks(rotation=45)
plt.show()
```

The bar chart represents the distribution of different categories of items purchased by customers. Each bar corresponds to a category, and the height of the bar indicates the number of occurrences of that category in the dataset.

```python
frequency_counts = df['Frequency of Purchases'].value_counts()
plt.bar(frequency_counts.index, frequency_counts.values)
plt.xlabel('Frequency of Purchases')
plt.ylabel('Count')
plt.title('Frequency of Purchases')
plt.xticks(rotation=45)  
plt.show()
```

The bar chart represents the frequency of purchases among customers.
